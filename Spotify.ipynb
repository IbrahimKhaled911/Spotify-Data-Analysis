```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('tracks.csv')
```

```python
most_populer=df.query('popularity>90', inplace = False).sort_values('popularity', ascending= False)
most_populer[:10]
```

```python
df.set_index('release_date', inplace= True)
df.index=pd.to_datetime(df.index)
df.head()
```

```python
df[['artists']].iloc[18]
```

```python
df['duration']=df['duration_ms'].apply(lambda x: round (x/1000))
df.drop('duration_ms', inplace= True, axis=1)
```

```python
df.duration.head()
```

```python
corr_df=df.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df=df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='loudness', x='energy', color= 'c').set(title='Loudness vs Energy Correlation')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='popularity', x='acousticness', color= 'c').set(title='popularity vs acousticness Correlation')
plt.show()
```

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
print(years)
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of Songs Per Year')
plt.show()
```

```python
total_dr=df.duration
fig_dims = (18,7)
fig, ax= plt.subplots(figsize=fig_dims)
fig= sns.barplot(x = years, y= total_dr, ax= ax, errwidth= False).set(title='Years vs Duration')
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr= df.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title='Years vs Duration')
plt.xticks(rotation=60)
plt.show()
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.title('Duration of the Songs in Defferent Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in milli seconds')
plt.ylabel('Genres')
plt.show()
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous= df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title='Top 5 Genres by Popularity')
plt.show()
```
